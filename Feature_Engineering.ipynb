## Feature Engineering

Análisis de Sentimientos

En el conjunto de datos user_reviews se incluyen reseñas de juegos hechos por distintos usuarios. Debes crear la columna 'sentiment_analysis' aplicando análisis de sentimiento con PNL con la siguiente escala: debes tomar el valor '0' si es malo, '1' si es neutral y '2' si es positivo. Esta nueva columna debe reemplazar la de user_reviews.review para facilitar el trabajo de los modelos de aprendizaje automático y el análisis de datos. De no ser posible este análisis por estar ausente la reseña escrita, debe tomar el valor de 1.

Importamos las librerias necesarias

```python
import pandas as pd
import pyarrow as pa
import ast
import numpy as np
import csv
from textblob import TextBlob
```

Extraemos el archivo CSV creado en el ETL user_reviews

```python
archivo = r"C:\Users\ayrc2\Documentos\Proyecto Individual DPT03\PI_ML_OPS_Steam_Games\user_reviews.csv"
user_reviews= pd.read_csv(archivo)
```

```python
#Funcion para el analisis de sentimientos "tener en cuenta que se debe importar la libreria TextBlob"

def analizar_sentimiento(review):
    if review is None:
        return 1 #Neutral
    analisis = TextBlob(review)
    polaridad = analisis.sentiment.polarity
    if polaridad > 0.1:
        return 2 #'Positivo'
    elif polaridad < -0.1:
        return 0 #'Negativo'
    else:
        return 1 #'Neutral'
```

```python
#Se aplica la función creada a la coliumna reviwes, para realizar el análisis y generaer la columna solicitada Sentiment_analysis
user_reviews ["sentiment_analisis"]= user_reviews["review"].astype(str).apply(analizar_sentimiento)
user_reviews.head(5)
```

```python
#Verificamos con una de las reseñas si el analisis de sentimientos es correcto

print(user_reviews["review"][1000])
user_reviews["sentiment_analisis"][1000]
```

```python
# Al verificar que la función devuelve el valor correcto procedemos a eliminar la columna review y dejamos sentiment_analisis

user_reviews = user_reviews.drop(columns=["review"])
user_reviews
```

```python
# Guardamos el nuevo archivo con la columna sentiment:analisis reemplazando el anterior csv

user_reviews = user_reviews.to_csv("user_reviews_final.csv", index=False)
```
